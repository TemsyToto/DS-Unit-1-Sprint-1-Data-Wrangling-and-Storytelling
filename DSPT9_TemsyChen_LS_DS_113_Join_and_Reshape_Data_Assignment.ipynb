```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

```python
# table2 -> tidy

#First I move the rows into a column of values with reset_index

table2 = table2.reset_index()
table2
```

```python
#using melt, I change the table

tidy = table2.melt(id_vars='index',value_vars=['John Smith','Jane Doe','Mary Johnson'])
tidy
```

```python
#I rename the columns for clarity

tidy_rename = tidy.rename(columns={
    'index':'trt',
    'variable':'name',
    'value':'result'
})
tidy_rename
```

```python
#clean up the treatment row by getting rid of the word "treatment"
tidy_rename.trt = tidy_rename.trt.str.replace('treatment','')
tidy_rename
```

```python
# tidy -> table2
#step 1, use the pivot_table function to reverse the melt
wide = tidy_rename.pivot_table(values='result',index='trt',columns='name')
wide
```

```python
#reorder the columns
wide2 = wide[['John Smith','Jane Doe','Mary Johnson']]
wide2
```

```python
#reset the index and rename the trt column to index.
wide3 = wide2.reset_index().rename(columns={
    'trt':'index',
})
wide3
```

```python
#Tried to delete "name" from the table, but couldn't figure it out.
#Also, how to add "treatment" back into a and b values?
wide4 = wide3.rename(index={
    'name':'diff'
})
wide4
```

## Join Data Stretch Challenge

The [Instacart blog post](https://tech.instacart.com/3-million-instacart-orders-open-sourced-d40d29ead6f2) has a visualization of "**Popular products** purchased earliest in the day (green) and latest in the day (red)." 

The post says,

> "We can also see the time of day that users purchase specific products.

> Healthier snacks and staples tend to be purchased earlier in the day, whereas ice cream (especially Half Baked and The Tonight Dough) are far more popular when customers are ordering in the evening.

> **In fact, of the top 25 latest ordered products, the first 24 are ice cream! The last one, of course, is a frozen pizza.**"

Your challenge is to reproduce the list of the top 25 latest ordered popular products.

We'll define "popular products" as products with more than 2,900 orders.



```python
##### YOUR CODE HERE #####

#First take the orders data frame and create a subset of all the orders taken after 12.

orders_subset = orders[orders['order_hour_of_day']>=12]
orders_subset['order_hour_of_day'].value_counts()
```

```python
#Merge the orders_subset with order_products

afternoon_orders = pd.merge(orders_subset, order_products)
afternoon_orders
```

```python
#Make another subset to shrink this dataframe

afternoon_orders_small = afternoon_orders.drop(['eval_set','order_dow','days_since_prior_order','add_to_cart_order','reordered','user_id'], axis=1)
afternoon_orders_small
```

```python
#combine this with the product_id dataframe

afternoon_orders_final = pd.merge(afternoon_orders_small, products)
afternoon_orders_final
```

```python
afternoon_orders_final2 = afternoon_orders_final.drop(['order_number','product_id','aisle_id','department_id'], axis=1)
afternoon_orders_final2
```

```python
#minimize the data again to only include items purchased in the last hour of the day
pd.set_option('display.max_rows',200)

afternoon_orders_24hr = afternoon_orders_final2[afternoon_orders_final2['order_hour_of_day']>=22]
afternoon_orders_24hr
```

```python
#I'm not getting the same list that the assigment is asking for, no ice cream or pizza.

afternoon_orders_24hr['product_name'].value_counts()[:25].sort_values(ascending=False)
```

```python
#Vignesh suggests using the groupby function to find the total amount of order of each product each hour

afternoon_orders_24hr['Counts'] = afternoon_orders_24hr.groupby(['product_name'])['product_name'].transform('counts')

# subset['Counts'] = subset.groupby(['product_name'])['product_name'].transform('count')
#Still working on this
```
